## About the dataset

   1. sepal length in cm
   2. sepal width in cm
   3. petal length in cm

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
HR_age = sorted(HR.Age)
q1 = np.percentile(HR_age, [25])
q3 = np.percentile(HR_age, [75])
IQR = q3 - q1
print("IQR - Interquartile range of Age = ", *IQR )
```
